```python
import csv
from ics import Calendar, Event
import pytz
import datetime as dt

with open('Colloscope MP.csv', newline='', encoding='UTF-8') as csvfile:
    spamreader = csv.reader(csvfile, delimiter=';')
    res = []
    for row in spamreader:
        res.append(row)

res = res[7:] #garde que les lignes avec les groupes et les horaires

liste = [[res[j][i] for j in range(len(res))] for i in range(len(res[0]))] #inverse lignes et colonnes


with open('Colles de français.csv', newline='', encoding='UTF-8') as csvfile:
    spamreader = csv.reader(csvfile, delimiter=';')
    res_fr = []
    for row in spamreader:
        res_fr.append(row)


dico_sem = {'Lu':0,'Ma':1,'Me':2,'Je':3,'Ve':4,'Sa':5}
dico_matiere = {'MATHS':'Colle de Mathématiques', 'ANGLAIS':"Colle d'Anglais", 'PHYSIQUE':'Colle de Physique','TP ITC':'TP ITC', 'TP Option Info':'TP Option Info','TP Option SI':'TP Option SI','Option Info':'Option Info','DS':'DS','TD Option Info':'TD Option Info','TD SI':'TD SI','SI-Cours':'SI-Cours', 'Option SI':'Option SI'}
dico_mois = {'sept':9,'oct':10,'janv.':1,'févr.':2,'mars':3,'avr.':4,'mai':5,'juin':6}
dico_heure = {'14h10':(14,10),'15h05':(15,5),'15h20':(15,20),'16h15':(16,15),'16h20':(16,20),'17h15':(17,15),'17h20':(17,20),'18h15':(18,15),'18h20':(18,20),'19h15':(19,15),'19h20':(19,20),'08h00':(8,0),'09h55':(9,55),'13h10':(13,10),'14h05':(14,5),'12h05':(12,5),'12h10':(12,10),'13h05':(13,5), '10h10':(10,10), '17h15':(17,15), '17h45':(17,45), '18h15':(18,15), '18h45':(18,45), '19h15':(19,15), '17h45':(17,45), '18h45':(18,45)}
dico_ds = {'Info':"DS d'option Info", 'maths':'DS de Mathématiques', 'physique':'DS de Physique','français':'DS de Français','anglais':"DS d'Anglais", 'SI':"DS d'option SI", 'ITC':"DS d'informatique pour tous"}
liste_conge = [dt.datetime(year=2022,month=10,day=i) for i in range(23,32)] + [dt.datetime(year=2022,month=11,day=i) for i in range(1,7)] + [dt.datetime(year=2022,month=12,day=i) for i in range(18,32)] + [dt.datetime(year=2023,month=1,day=i) for i in range(1,3)] + [dt.datetime(year=2023,month=2,day=i) for i in range(19,29)] + [dt.datetime(year=2023,month=3,day=i) for i in range(1,6)] + [dt.datetime(year=2023,month=4,day=i) for i in range(23,31)] + [dt.datetime(year=2023,month=5,day=i) for i in range(1,9)] + [dt.datetime(year=2023,month=11,day=11)]

liste_groupe_colle = [('ELEVE_1', 'ELEVE_2', 'ELEVE_3'),
                      ('ELEVE_4', 'ELEVE_5', 'ELEVE_6'),
                      ('ELEVE_7', 'ELEVE_8', 'ELEVE_9'),
                      ('ELEVE_10', 'ELEVE_11', 'ELEVE_12'),
                      ('ELEVE_13', 'ELEVE_14', 'ELEVE_15'),
                      ('ELEVE_16', 'ELEVE_17', 'ELEVE_18'),
                      ('ELEVE_19', 'ELEVE_20', 'ELEVE_21'),
                      ('ELEVE_22', 'ELEVE_23', 'ELEVE_24'),
                      ('ELEVE_25', 'ELEVE_26', 'ELEVE_27'),
                      ('ELEVE_28', 'ELEVE_29', 'ELEVE_30'),
                      ('ELEVE_31', 'ELEVE_32', 'ELEVE_33'),
                      ('ELEVE_34', 'ELEVE_35', 'ELEVE_36')]

liste_groupe_info = ['ELEVE_1', 'ELEVE_2', 'ELEVE_3', 'ELEVE_4', 'ELEVE_5', 'ELEVE_6', 'ELEVE_7', 'ELEVE_8', 'ELEVE_9', 'ELEVE_10', 'ELEVE_11', 'ELEVE_12', 'ELEVE_16', 'ELEVE_17', 'ELEVE_18', 'ELEVE_19', 'ELEVE_20', 'ELEVE_21']
liste_groupe_si = ['ELEVE_13', 'ELEVE_14', 'ELEVE_15', 'ELEVE_22', 'ELEVE_23', 'ELEVE_24', 'ELEVE_25', 'ELEVE_26', 'ELEVE_27', 'ELEVE_28', 'ELEVE_29', 'ELEVE_30', 'ELEVE_31', 'ELEVE_32', 'ELEVE_33', 'ELEVE_34', 'ELEVE_35', 'ELEVE_36']

liste_groupe_tipe_info = ['ELEVE_10', 'ELEVE_13', 'ELEVE_8', 'ELEVE_11', 'ELEVE_17', 'ELEVE_32', 'ELEVE_26']
liste_groupe_tipe_maths = ['ELEVE_1', 'ELEVE_4', 'ELEVE_2', 'ELEVE_5', 'ELEVE_14', 'ELEVE_3', 'ELEVE_9', 'ELEVE_12', 'ELEVE_15']
liste_groupe_tipe_physique = ['ELEVE_7', 'ELEVE_16', 'ELEVE_19', 'ELEVE_23', 'ELEVE_25', 'ELEVE_28', 'ELEVE_31', 'ELEVE_34', 'ELEVE_20', 'ELEVE_22', 'ELEVE_29', 'ELEVE_35', 'ELEVE_6', 'ELEVE_18', 'ELEVE_21', 'ELEVE_24', 'ELEVE_27', 'ELEVE_30', 'ELEVE_33', 'ELEVE_36']

liste_classe = liste_groupe_info + liste_groupe_si
```

```python
def clean(l):
    """
        tranforme (['MATHS', 'M. PROF_1', 'Ve 17h20-18h20', '225'], '15\n31 janv.')
        en liste de Event()
    """
    res = []
    for i in l:
        e = Event()
        e.name = dico_matiere[i[0][0]]
        e.description = 'MP\n'+dico_matiere[i[0][0]] + ' - ' + i[0][1]
        e.location = i[0][3]
        date_i = dt.date.fromisoformat(i[1])
        jour = date_i.day
        mois = date_i.month
        annee = date_i.year
        decalage = dt.datetime(year=annee,month=mois,day=jour) < dt.datetime(2022,10,27) or dt.datetime(year=annee,month=mois,day=jour) > dt.datetime(2023,3,26)
        e.begin = str(dt.datetime(year=annee,month=mois, day=jour,hour=dico_heure[i[0][2][-11:-6]][0],minute=dico_heure[i[0][2][-11:-6]][1],tzinfo=pytz.timezone('Etc/GMT'))+dt.timedelta(days=dico_sem[i[0][2][:2]], hours=-1-decalage))
        e.end = str(dt.datetime(year=annee,month=mois, day=jour,hour=dico_heure[i[0][2][-5:]][0],minute=dico_heure[i[0][2][-5:]][1],tzinfo=pytz.timezone('Etc/GMT'))+dt.timedelta(days=dico_sem[i[0][2][:2]], hours=-1-decalage))
        res.append(e)
    return res

def liste_colles(groupe): #renvoie toutes les colles d'un groupe
    l=[]
    for jour in range(len(liste)):
        for j in range(len(liste[jour])):
            if liste[jour][j] == str(groupe):
                l.append((res[j][:4],liste[jour][0]))
    return l

def colle_fr(nom): #renvoie la colle de français en tant qu'événement
    i_date = 2
    for d in res_fr[2:]:
        i_heure = 1
        for p in d[1:]:
            if p[p.index(' ')+1:].upper() == nom.upper():
                e = Event()
                e.name = 'Colle de Français'
                e.description = 'MP\nColle de Français - Mme PROF_3'
                e.location = '320'
                date_i = dt.date.fromisoformat(res_fr[i_date][0])
                jour = date_i.day
                mois = date_i.month
                annee = date_i.year
                decalage = dt.datetime(year=annee,month=mois,day=jour) < dt.datetime(2022,10,27) or dt.datetime(year=annee,month=mois,day=jour) > dt.datetime(2023,3,26)
                e.begin = str(dt.datetime(year=annee,month=mois, day=jour,hour=dico_heure[res_fr[0][i_heure]][0],minute=dico_heure[res_fr[0][i_heure]][1],tzinfo=pytz.timezone('Etc/GMT'))+dt.timedelta(hours=-1-decalage))
                e.end = str(dt.datetime(year=annee,month=mois, day=jour,hour=dico_heure[res_fr[1][i_heure]][0],minute=dico_heure[res_fr[1][i_heure]][1],tzinfo=pytz.timezone('Etc/GMT'))+dt.timedelta(hours=-1-decalage))
                return e
            i_heure += 1
        i_date += 1

def semaine_normale(lundi): #lundi au format dt.datetime(...) et retourne une liste de Event()
    res = []
    jour = lundi.day
    mois = lundi.month
    annee = lundi.year
    decalage = dt.datetime(year=annee,month=mois,day=jour) < dt.datetime(2022,10,27) or dt.datetime(year=annee,month=mois,day=jour) > dt.datetime(2023,3,26)
    for m in [(0,(10,10),(12,5)),(2,(13,10),(15,5)),(3,(8,0),(9,55)),(3,(10,10),(12,5)),(4,(10,10),(12,5))]:
        e = Event()
        e.description = 'MP\nMathématiques - M. PROF_1'
        e.name = 'Mathématiques'
        e.begin = str(dt.datetime(year=annee,month=mois, day=jour,hour=m[1][0],minute=m[1][1],tzinfo=pytz.timezone('Etc/GMT'))+dt.timedelta(days=m[0], hours=-1-decalage))
        e.end = str(dt.datetime(year=annee,month=mois, day=jour,hour=m[2][0],minute=m[2][1],tzinfo=pytz.timezone('Etc/GMT'))+dt.timedelta(days=m[0], hours=-1-decalage))
        e.location = '111'
        if dt.datetime(year=annee,month=e.begin.month,day=e.begin.day) not in liste_conge:
            res.append(e)
    for p in [(1,(8,0),(9,55)),(1,(14,5),(16,15)),(2,(8,0),(9,55)),(4,(8,0),(9,55))]:
        e = Event()
        e.description = 'MP\nPhysique - M. PROF_2'
        e.name = 'Physique'
        e.begin = str(dt.datetime(year=annee,month=mois, day=jour,hour=p[1][0],minute=p[1][1],tzinfo=pytz.timezone('Etc/GMT'))+dt.timedelta(days=p[0], hours=-1-decalage))
        e.end = str(dt.datetime(year=annee,month=mois, day=jour,hour=p[2][0],minute=p[2][1],tzinfo=pytz.timezone('Etc/GMT'))+dt.timedelta(days=p[0], hours=-1-decalage))
        e.location = '111'
        if dt.datetime(year=annee,month=e.begin.month,day=e.begin.day) not in liste_conge:
            res.append(e)
    e = Event()
    e.description = 'MP\nFrançais Philosophie - Mme PROF_3'
    e.name = 'Français Philosophie'
    e.begin = str(dt.datetime(year=annee,month=mois, day=jour,hour=10,minute=10,tzinfo=pytz.timezone('Etc/GMT'))+dt.timedelta(days=2, hours=-1-decalage))
    e.end = str(dt.datetime(year=annee,month=mois, day=jour,hour=12,minute=5,tzinfo=pytz.timezone('Etc/GMT'))+dt.timedelta(days=2, hours=-1-decalage))
    e.location = '111'
    if dt.datetime(year=annee,month=e.begin.month,day=e.begin.day) not in liste_conge:
        res.append(e)

    e = Event()
    e.description = 'MP\nInformatique pour tous - M. PROF_4'
    e.name = 'Informatique pour tous'
    e.begin = str(dt.datetime(year=annee,month=mois, day=jour,hour=12,minute=10,tzinfo=pytz.timezone('Etc/GMT'))+dt.timedelta(days=1, hours=-1-decalage))
    e.end = str(dt.datetime(year=annee,month=mois, day=jour,hour=13,minute=5,tzinfo=pytz.timezone('Etc/GMT'))+dt.timedelta(days=1, hours=-1-decalage))
    e.location = '111'
    if dt.datetime(year=annee,month=e.begin.month,day=e.begin.day) not in liste_conge and dt.datetime(year=annee,month=e.begin.month,day=e.begin.day) < dt.datetime(2023,1,30):
        res.append(e)

    e = Event()
    e.description = 'MP\nAnglais - M. PROF_5'
    e.name = 'Anglais'
    e.begin = str(dt.datetime(year=annee,month=mois, day=jour,hour=10,minute=10,tzinfo=pytz.timezone('Etc/GMT'))+dt.timedelta(days=1, hours=-1-decalage))
    e.end = str(dt.datetime(year=annee,month=mois, day=jour,hour=12,minute=5,tzinfo=pytz.timezone('Etc/GMT'))+dt.timedelta(days=1, hours=-1-decalage))
    e.location = '111'
    if dt.datetime(year=annee,month=e.begin.month,day=e.begin.day) not in liste_conge:
        res.append(e)
    return res

def td(groupe, option, lundi):#groupe 1 = physique en premier, option : info ou si, lundi au format dt.datetime(...)
    if option == 'B':
        option = 2
    else:
        option = 1
    res = []
    jour = lundi.day
    mois = lundi.month
    annee = lundi.year
    decalage = dt.datetime(year=annee,month=mois,day=jour) < dt.datetime(2022,10,27) or dt.datetime(year=annee,month=mois,day=jour) > dt.datetime(2023,3,26)
    horraire = [None,((13,10),(15,5)),((15,20),(17,15))]
    e1 = ['MP\nTP de Physique - M. PROF_2','TP-Physique',str(dt.datetime(year=annee,month=mois, day=jour,hour=horraire[groupe][0][0],minute=horraire[groupe][0][1],tzinfo=pytz.timezone('Etc/GMT'))+dt.timedelta(days=4,hours=-1-decalage)),
          str(dt.datetime(year=annee,month=mois, day=jour,hour=horraire[groupe][1][0],minute=horraire[groupe][1][1],tzinfo=pytz.timezone('Etc/GMT'))+dt.timedelta(days=4,hours=-1-decalage)),'314']
    e2 = ['MP\nTD de Mathématiques - M. PROF_1','TD-Mathématiques',str(dt.datetime(year=annee,month=mois, day=jour,hour=horraire[3-option][0][0],minute=horraire[3-option][0][1],tzinfo=pytz.timezone('Etc/GMT'))+dt.timedelta(hours=-1-decalage)),
          str(dt.datetime(year=annee,month=mois, day=jour,hour=horraire[3-option][1][0],minute=horraire[3-option][1][1],tzinfo=pytz.timezone('Etc/GMT'))+dt.timedelta(hours=-1-decalage)),'111']
    for e in (e1,e2):
        ev = Event()
        ev.description = e[0]
        ev.name = e[1]
        ev.begin = e[2]
        ev.end = e[3]
        ev.location = e[4]
        if dt.datetime(year=annee,month=ev.begin.month,day=ev.begin.day) not in liste_conge:
            res.append(ev)
    return res


def tipe(groupe,lundi): #groupe = le groupe de tipe (info=1, maths=2, physique=3), lundi au format dt.datetime(...)
    res = []
    jour = lundi.day
    mois = lundi.month
    annee = lundi.year
    decalage = dt.datetime(year=annee,month=mois,day=jour) < dt.datetime(2022,10,27) or dt.datetime(year=annee,month=mois,day=jour) > dt.datetime(2023,3,26)
    if groupe == 1:#info
        e = Event()
        e.description = "MP\nTIPE d'Informatique - M. PROF_6"
        e.name = 'TIPE-Informatique'
        e.begin = str(dt.datetime(year=annee,month=mois, day=jour,hour=16,minute=20,tzinfo=pytz.timezone('Etc/GMT'))+dt.timedelta(days=3, hours=-1-decalage))
        e.end = str(dt.datetime(year=annee,month=mois, day=jour,hour=17,minute=15,tzinfo=pytz.timezone('Etc/GMT'))+dt.timedelta(days=3, hours=-1-decalage))
        e.location = '115'
        if dt.datetime(year=annee,month=e.begin.month,day=e.begin.day) not in liste_conge:
            res.append(e)
    if groupe == 2:#maths
        e = Event()
        e.description = "MP\nTIPE de Mathématiques - M. PROF_1"
        e.name = 'TIPE-Mathématiques'
        e.begin = str(dt.datetime(year=annee,month=mois, day=jour,hour=13,minute=10,tzinfo=pytz.timezone('Etc/GMT'))+dt.timedelta(days=4, hours=-1-decalage))
        e.end = str(dt.datetime(year=annee,month=mois, day=jour,hour=15,minute=5,tzinfo=pytz.timezone('Etc/GMT'))+dt.timedelta(days=4, hours=-1-decalage))
        e.location = '111'
        if dt.datetime(year=annee,month=e.begin.month,day=e.begin.day) not in liste_conge:
            res.append(e)
    if groupe == 3:#physique
        e = Event()
        e.description = "MP\nTIPE de Physique - M. PROF_2"
        e.name = 'TIPE-Physique'
        e.begin = str(dt.datetime(year=annee,month=mois, day=jour,hour=15,minute=20,tzinfo=pytz.timezone('Etc/GMT'))+dt.timedelta(days=3, hours=-1-decalage))
        e.end = str(dt.datetime(year=annee,month=mois, day=jour,hour=17,minute=15,tzinfo=pytz.timezone('Etc/GMT'))+dt.timedelta(days=3, hours=-1-decalage))
        e.location = '411'
        if dt.datetime(year=annee,month=e.begin.month,day=e.begin.day) not in liste_conge:
            res.append(e)
    return res

def leven(mot1, mot2):
    n = len(mot1)
    m = len(mot2)
    A = [[i + j for j in range(m + 1)] for i in range(n + 1)]
    for i in range(n):
        for j in range(m):
            A[i + 1][j + 1] = min(A[i][j + 1] + 1, A[i + 1][j] + 1, A[i][j] + int(mot1[i] != mot2[j]))
    return A[n][m]

def min_leven(mot, liste):
    return min([(leven(mot, i), i) for i in liste_classe])[1]

def nom_to_groupe(nom): #renvoie le groupe_spe (info='B', si='A'), le groupe de tipe (info=1, maths=2, physique=3), le groupe de colle, groupe de td
    res = []
    if nom in liste_groupe_info:
        res.append('B')
    elif nom in liste_groupe_si:
        res.append('A')
    if nom in liste_groupe_tipe_info:
        res.append(1)
    elif nom in liste_groupe_tipe_maths:
        res.append(2)
    elif nom in liste_groupe_tipe_physique:
        res.append(3)
    c = 1
    for g in liste_groupe_colle:
        if nom in g:
            res.append(c)
            break
        c += 1
    res.append(2 - (c in [7, 8, 9, 10, 11, 12, 13]))
    return res


def create_semaine(lundi,groupe_td,groupe_tipe,groupe_spe): # liste de Event() sans les options
    res = []
    for ev in semaine_normale(lundi):
        res.append(ev)
    for t in td(groupe_td, groupe_spe, lundi):
        res.append(t)
    for ti in tipe(groupe_tipe,lundi):
        res.append(ti)
    return res
```

```python
def create_calendrier_colles(nom): #groupe = groupe de colles
    groupe_spe, groupe_tipe, groupe_colle, groupe_td = nom_to_groupe(nom)
    c = Calendar()
    for ev in clean(liste_colles(groupe_colle)):
        c.events.add(ev)
    c.events.add(colle_fr(nom))
    with open('Agenda_colles.ics', 'w') as f:
        f.write(str(c))
    return c

def create_agenda(nom): #groupe_td (1 = physique en premier), groupe_tipe (1 = physique en premier), groupe_spe='A'(SI), 'B'(Info)
    groupe_spe, groupe_tipe, groupe_colle, groupe_td = nom_to_groupe(nom)
    c=Calendar()
    for s in clean(liste_colles(groupe_spe)): #ajout tp info
        c.events.add(s)
    for i in range(28):
        lundi = dt.datetime(year=2022,month=9,day=19)+dt.timedelta(days=7)*i
        for j in create_semaine(lundi,groupe_td,groupe_tipe,groupe_spe):
            c.events.add(j)
    for ds in ('maths','physique','français','anglais','SI / Info','ITC'):
        for mat in clean(liste_colles(ds)):
            if ds == 'SI / Info':
                if groupe_spe == 'B':
                    mat.name = dico_ds['Info']
                else:
                    mat.name = dico_ds['SI']
            else:
                mat.name = dico_ds[ds]
            mat.description = 'Devoir Surveillé'+mat.name[2:]
            c.events.add(mat)

    with open('Agenda_cours.ics', 'w') as f:
        f.write(str(c))
    return c
```
